```python
# Filling null values
# with 0
MetricsRatios_df.fillna(value = 0,
          inplace = True)
 
# Show the DataFrame
print(MetricsRatios_df)
```

```python
plt.figure(figsize=(12,6))    # Types of sectors and the topmost sector amongst all companies.
data = MetricsRatios_df.sector.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
plt.show()
```

```python
plt.figure(figsize=(12,6))     # Types of subsectors and the topmost subsector amongst all companies.
sns.countplot(x='subsector', data=MetricsRatios_df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
data = MetricsRatios_df.sector.value_counts()  # sector percentage


labels = (np.array(data.index))
sizes = (np.array((data / data.sum())*100))
colors = ['red', 'lightskyblue','pink','yellow','orange','green','silver']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("sector percentage")
plt.show()
          
```

```python
plt.figure(figsize=(20, 6))
sns.histplot(MetricsRatios_df['sector'], kde=True)

plt.title('Distribution of Sectors')
plt.xlabel('Sector')
plt.ylabel('count')

plt.show()
```
